## Этап 1. Подготовка данных

```python
# импортируем библиотеки и модули

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler, OrdinalEncoder 
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error, accuracy_score, confusion_matrix, precision_score, recall_score 
from sklearn.metrics import roc_auc_score, f1_score, classification_report, roc_curve
from sklearn.utils import shuffle

import warnings
warnings.filterwarnings('ignore')
np.random.seed(12345)
```

## Этап 2. Исследование задачи

### Этап 2.1 Изучение классов

```python
# посмотрим как распределены классы в целевом признаке

data.exited.value_counts()
```

Выявлен дисбаланс классов: в имеющихся данных клиенты уходили из банка в 4 раза реже, чем оставались. Если мы будем всех людей прогнозировать, что они не уйдут из банка, то в 80% случаев мы будем правы.

```python
# посмотрим визуально есть ли корреляция между признаками клиентов и распределением целевого признака

data_graph = data[['credit_score', 'age', 'tenure', 'balance', 'exited']].query('tenure != -1')
sns.pairplot(data=data_graph, hue ='exited')
plt.show();
```

Линейной зависимости (корреляции) между признаками не наблюдается.

### Этап 2.2 Деление на выборки

Для проведения исследования и обучения модели и оценки её эффективности разобьем имеющиеся данные на три части: обучающую, валидационную и тестовую выборки. Размеры тестового и валидационного наборов обычно равны. Исходные данные разбиваем в соотношении 3:1:1 (60% / 20% / 20%). Также используем параметр stratify для равномерного распределения классов (1 и 0 в столбце exited) в каждом наборе.
- data_train - обучающая выборка 60%
- data_valid - валидационная выборка 20%
- data_test - тестовая выборка 20%

```python
# делим данные на 3 выборки

data_train, data_check = train_test_split(data, test_size=0.4, random_state=12345, stratify=data.exited)
data_test, data_valid = train_test_split(data_check, test_size=0.5, random_state=12345, stratify=data_check.exited)
```

```python
# выведем на экран созданные выборки

print(f'Размер обучающей выборки: {data_train.shape}')
print(f'Размер валидационной выборки: {data_valid.shape}')
print(f'Размер тестовой выборки: {data_test.shape}')
```

### Этап 2.3 Определение признаков, целевого признака. Масштабирование данных

Создадим переменные для трех выборок, выделив признаки, на которых модель будет обучаться, и целевой признак, т.е. факт ухода клиента из банка.

```python
# создадим переменные для признаков (features) и целевого признака (target)

features_train = data_train.drop(['exited'], axis=1)
target_train = data_train['exited']

features_valid = data_valid.drop(['exited'], axis=1)
target_valid = data_valid['exited']

features_test = data_test.drop(['exited'], axis=1)
target_test = data_test['exited']
```

```python
# выведем на экран размеры таблиц, которые хранятся в новых переменных 

print(features_train.shape)
print(target_train.shape)
print(features_valid.shape)
print(target_valid.shape)
print(features_test.shape)
print(target_test.shape)
```

```python
# масштабируем числовые данные в столбцах из списка

numeric = ['credit_score', 'age', 'tenure', 'balance', 'estimated_salary']
ss = StandardScaler()
ss.fit(features_train[numeric])
features_train[numeric] = ss.transform(features_train[numeric])
features_valid[numeric] = ss.transform(features_valid[numeric])
features_test[numeric] = ss.transform(features_test[numeric])

# смотрим, что получилось

display(features_train.head())
display(features_valid.head())
display(features_test.head())
```

### Этап 2.4 Преобразование категориальных признаков техникой One-Hot Encoding, OHE

Логистическая регрессия, которая используется для решения задач классификации, не подходит, если данные содержат категориальные признаки. У нас это столбцы geography и gender, а также num_of_products, has_cr_card, is_active_member.

Преобразуем категориальные признаки из обучающей выборки в численные техникой One-Hot Encoding, OHE. Один из вновь созданных столбцов по каждому признаку удалим с помощью аргумента drop_first.

```python
# преобразуем категориальные столбцы в численные в обучающей выборке с помощью метода get_dummies

features_train_ohe = pd.get_dummies(features_train, columns = ['geography', 'gender', 
                                                               'num_of_products', 'has_cr_card', 
                                                               'is_active_member'], drop_first=True)
features_train_ohe = features_train_ohe.drop(['row_number', 'customer_id', 'surname'], axis=1)

display(features_train_ohe.head())
```

```python
# преобразуем категориальные столбцы в численные в валидационной выборке с помощью метода get_dummies

features_valid_ohe = pd.get_dummies(features_valid, columns = ['geography', 'gender', 
                                                               'num_of_products', 'has_cr_card', 
                                                               'is_active_member'], drop_first=True)
features_valid_ohe = features_valid_ohe.drop(['row_number', 'customer_id', 'surname'], axis=1)

display(features_valid_ohe.head())
```

```python
# преобразуем категориальные столбцы в численные в тестовой выборке с помощью метода get_dummies

features_test_ohe = pd.get_dummies(features_test, columns = ['geography', 'gender', 
                                                               'num_of_products', 'has_cr_card', 
                                                               'is_active_member'], drop_first=True)
features_test_ohe = features_test_ohe.drop(['row_number', 'customer_id', 'surname'], axis=1)

display(features_test_ohe.head())
```

### Этап 2.5 Обучение модели на алгоритме логистической регрессии

```python
# обучим модель на алгоритме логистической регрессии

lr = LogisticRegression()
lr_parameters = {'penalty' : ['l1', 'l2', 'elasticnet'],
                 'C' : np.logspace(-3,3,5), 
                 'solver' : ['newton-cg', 'lbfgs', 'liblinear']}
clf_lr = GridSearchCV(lr, lr_parameters, n_jobs=2, verbose=1, cv=5, scoring= 'f1')
%time clf_lr.fit(features_train_ohe, target_train)
display(clf_lr.best_params_)
display(clf_lr.best_score_)
```

```python
# проверим точность модели с наилучшими параметрами на валидационной выборке

model_lr = LogisticRegression(C = 1.0, 
                            penalty = 'l2', 
                            solver = 'liblinear')
%time model_lr.fit(features_train_ohe, target_train)
model_lr_pred = model_lr.predict(features_valid_ohe)
result_lr = f1_score(target_valid, model_lr_pred)
print(f'F1 лучшей модели на валидационной выборке: {result_lr}')
```

Модель, обученная на алгоритме логистической регрессии, даже после перебора некоторых гиперпараметров показала довольно низкое значение F1-меры = 0.49.

### Этап 2.6 Преобразование категориальных признаков техникой Ordinal Encoding, OE

Чтобы обучить модель на алгоритме решающего дерева или случайного леса категориальные признаки также надо будет преобразовать в численные. Для этого применяется другая техника - Ordinal Encoding, OE.

```python
# преобразуем категориальные столбцы в численные в обучающей выборке

oe = OrdinalEncoder()
features_train_oe = pd.DataFrame(oe.fit_transform(features_train), columns=features_train.columns)

display(features_train_oe.head())
```

```python
# преобразуем категориальные столбцы в численные в валидационной выборке

features_valid_oe = pd.DataFrame(oe.fit_transform(features_valid), columns=features_valid.columns)

display(features_valid_oe.head())
```

```python
# преобразуем категориальные столбцы в численные в тестовой выборке

features_test_oe = pd.DataFrame(oe.fit_transform(features_test), columns=features_test.columns)

display(features_test_oe.head())
```

### Этап 2.7 Обучение модели на алгоритме решающего дерева

Теперь обучим модель решающего дерева и посчитаем значение F1-меры.

```python
# обучим модель на алгоритме решающего дерева

dtc = DecisionTreeClassifier()
dtc_parameters = {'max_depth': range (1,11)}
clf_dtc = GridSearchCV(dtc, dtc_parameters, n_jobs=2, verbose=1, scoring='f1')
%time clf_dtc.fit(features_train_oe, target_train)
display(clf_dtc.best_params_)
display(clf_dtc.best_score_)
```

```python
# проверим точность модели с наилучшими параметрами на валидационной выборке

model_dtc = DecisionTreeClassifier(max_depth = 5)
%time model_dtc.fit(features_train_oe, target_train)
model_dtc_pred = model_dtc.predict(features_valid_oe)
result_dtc = f1_score(target_valid, model_dtc_pred)
print(f'F1 лучшей модели на валидационной выборке: {result_dtc}')
```

Модель, обученная на алгоритме решающего дерева, также показала довольно низкое значение F1-меры = 0.49.

### Этап 2.8 Обучение модели на алгоритме случайного леса

А теперь обучим модель на случайном лесе и посчитаем значения F1-меры и еще некоторых метрик.

```python
# обучим модель на алгоритме случайного леса

rfc = RandomForestClassifier()
rfc_parameters = {'n_estimators': range (10, 31, 10),
                  'max_depth': range (1,11),
                 'min_samples_leaf': range (1,4),
                 'min_samples_split': range (2,5,2)}
clf_rfc = GridSearchCV(rfc, rfc_parameters, n_jobs=2, verbose=1, cv=5, scoring='f1')
%time clf_rfc.fit(features_train_oe, target_train)
display(clf_rfc.best_params_)
display(clf_rfc.best_score_)
```

## Этап 3. Борьба с дисбалансом

Чтобы увеличить качество модели при дисбалансе классов применим следующие техники: **upsampling и downsampling**. Настроим метрики классификации, учитывая вероятности классов. Построим ROC-кривую и вычислим площадь под ней.

В дальнейшем исследовании будем пользоваться масштабированными и преобразованными в технике One-Hot Encoding данными (из ячеек 21-23):
- features_train_ohe, target_train
- features_valid_ohe, target_valid
- features_test_ohe, target_test

### Этап 3.1 Обучение модели на алгоритме логистической регрессии, техники upsampling и downsampling

Сначала воспользуемся техникой **upsampling**: в обучающей выборке после разделения на отрицательные и положительные классы, увеличим положительный класс путем копирования имеющихся ответов. Затем обучим модель на логистической регрессии.

```python
# создадим функцию для копирования положительных классов и перемешивания, 
# затем зададим параметры для получения новых выборок,
# затем обучим модель на параметрах, взятых для лучшей модели из предыдущего 2 этапа

def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_upsampled = pd.concat([features_zeros] + [features_ones] * repeat)
    target_upsampled = pd.concat([target_zeros] + [target_ones] * repeat)
    
    features_upsampled, target_upsampled = shuffle(
        features_upsampled, target_upsampled, random_state=12345)
    
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train_ohe, target_train, 4)

model_lr_upsm = LogisticRegression(solver='liblinear', C = 1.0, penalty = 'l2', random_state=12345)
model_lr_upsm.fit(features_upsampled, target_upsampled)
predicted_valid_lr_upsm = model_lr_upsm.predict(features_valid_ohe)
print("F1 на валидационной выборке:", f1_score(target_valid, predicted_valid_lr_upsm))
```

Значение f1 удалось поднять с 0.4886 до 0.5575.

```python
# построим ROC-кривую

probabilities_valid_lr_upsm = model_lr_upsm.predict_proba(features_valid_ohe)
probabilities_one_valid_lr_upsm = probabilities_valid_lr_upsm[:, 1]

fpr, tpr, thresholds = roc_curve(target_valid, probabilities_one_valid_lr_upsm)

plt.figure()
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая модели логистической регрессии с техникой upsampling')
plt.show()

print()
auc_roc_lr_upsm = roc_auc_score(target_valid, probabilities_one_valid_lr_upsm)
print(f'AUC-ROC модели логистической регрессии с техникой upsampling: {auc_roc_lr_upsm}')
```

Теперь попробуем технику **downsampling**: в обучающей выборке после разделения на отрицательные и положительные классы, из отрицательных классов вычтем случайным образом часть ответов. Затем обучим модель на логистической регрессии.

```python
# создадим функцию для уменьшения отрицательных классов и перемешивания, 
# затем зададим параметры для получения новых выборок,
# затем обучим модель на параметрах, взятых для лучшей модели из предыдущего 2 этапа

def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled

features_downsampled, target_downsampled = downsample(features_train_ohe, target_train, 0.25)

model_lr_dnsm = LogisticRegression(solver='liblinear', C = 1.0, penalty = 'l2', random_state=12345)
model_lr_dnsm.fit(features_downsampled, target_downsampled)
predicted_valid_lr_dnsm = model_lr_dnsm.predict(features_valid_ohe)
print("F1 на валидационной выборке:", f1_score(target_valid, predicted_valid_lr_dnsm))
```

Значение f1 удалось поднять с 0.4886 до 0.5617. И это выше, чем при использовании техники upsampling. Но всё же недостаточно для требуемого результата (>=0.59).

```python
# построим ROC-кривую

probabilities_valid_lr_dnsm = model_lr_dnsm.predict_proba(features_valid_ohe)
probabilities_one_valid_lr_dnsm = probabilities_valid_lr_dnsm[:, 1]

fpr, tpr, thresholds = roc_curve(target_valid, probabilities_one_valid_lr_dnsm)

plt.figure()
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-кривая модели логистической регрессии с техникой downsampling')
plt.show()

print()
auc_roc_lr_dnsm = roc_auc_score(target_valid, probabilities_one_valid_lr_dnsm)
print(f'AUC-ROC модели логистической регрессии с техникой downsampling: {auc_roc_lr_dnsm}')
```

AUC-ROC моделей логистической регрессии с применением обеих техник (upsampling и downsampling) практически одинаковый и равен 0.83. Модель неплохо обучилась и хорошо предсказывает положительные классы.
