### 2.3 Обработка пропусков

```python
# посмотрим сколько пропусков в каждом столбце

data.isna().sum()
```

Отсутствуют данные по двум строкам в столбцах name и genre. Скорее всего это одни и те же две строки, и жанр игры не определен из-за отсутствия названия игры. В столбце year_of_release 269 пропусков. В строках с пропущенным наименованием игры также отсутствуют наименования жанров, оценки критиков и пользователей, а также рейтинги esrb. Удалим эти строки в виду их малоинформативности.

Поскольку 269 строк составляют 1,6% от всего объема данных и некоторые строки также имеют пропуски в оценках критиков и пользователей и рейтингов, то удалим эти строки.

```python
# удалим строки с пропусками в столбце year_of_release

data = data.dropna(subset=['year_of_release'])
```

```python
# проверим тип данных в столбце year_of_release

data.year_of_release.dtype
```

```python
# заменим тип на целочисленный

data.year_of_release = data.year_of_release.astype(int)
```

```python
# проверим снова тип данных в столбце year_of_release

data.year_of_release.dtype
```

```python
# посмотрим краткую информацию о столбце critic_score

data.critic_score.describe().round(2)
```

Похоже, что оценки критиков целочисленные. Можно заполнить недостающие значения нехарактерным значением ("заглушкой")  и изменить тип данных на целочисленный. 
Отсутствующих данных слишком много, чтобы удалять их или заполнять другими синтетическими данными, а "заглушку" мы сможем увидеть на графике или отфильтровать её. 

Как показывает опыт, оценки критиков (подразумевающие бОльшую объективность) и пользователей (более субъективные) не всегда коррелируют. Чтобы это проверить нужно будет расчитать коэффициент корреляции Пирсона после обработки пропусков. 

```python
# заполним пропуски в столбце critic_score на отрицательное число -1, чтобы потом поменять тип данных на целочисленный

data.critic_score = data.critic_score.fillna(value=-1)
```

```python
# заменим тип данных на целочисленный

data.critic_score = data.critic_score.astype(int)
```

Обработаем пропуски в столбце user_score.

```python
# посмотрим какие уникальные значения и с какой частотой встречаются в столбце user_score

data.user_score.sort_values().unique()
```

Особенностью этих данных является их строковый тип. Превалирует оценка tbd, которая расшифровывается как To Be Determined, то есть "Будет определено". Очень часто подобное сокращение можно встретить в анонсах компьютерных игр в графе "Дата выхода". Также эта аббревиатура может быть использована и в случае с уже вышедшей игрой, если у неё готовится к релизу какое-либо дополнение/модификация, однако пока не известно, когда именно оно будет выпущено. Таким образом получается, что оценка пользователей tbd означает, что игра пока не оценена/будет оценена/оценивается и на данный момент не несёт в себе балльной оценки. Лучше всего будет преобразовать этот тип данных в целочисленный. Но сначала заполним пропуски и изменим значение tbd на ноль.

```python
# заполним пропуски в столбце user_score на отрицательное число -1, чтобы потом поменять тип данных на целочисленный

data.user_score = data.user_score.fillna(value=-1)
```

```python
# изменим значения tbd на ноль, т.к. это значение не несет в себе никакой оценки

data.loc[data['user_score'] == 'tbd', 'user_score'] = 0
```

```python
# проверим заполнились ли столбцы

data.user_score.describe(include='all')
```

```python
# поменяем тип данных на числа с плавающей точкой

data.user_score = data.user_score.astype(float)
```

```python
# проверим уникальные значения и тип данных

data.user_score.sort_values().unique()
```

По данным ru.wikipedia.org **Entertainment Software Rating Board (ESRB)** - негосударственная организация, основное направление деятельности которой принятие и определение рейтингов для компьютерных видеоигр и другого развлекательного программного обеспечения в США и Канаде. 

Рейтинг игр ESRB основан на их содержании аналогично рейтинговым системам кинофильмов. Рейтинг игры напечатан на упаковке, содержится в её рекламе и указан на сайте игры. Компании не обязаны предоставлять игру на определение рейтинга до начала официальных продаж, так как её содержимое может значительно измениться в ходе разработки.

Рейтинг состоит из двух частей: знака рейтинга и краткого описания содержимого. Знак рейтинга обычно располагается в нижнем левом или правом углу на лицевой стороне упаковки и определяет наиболее подходящую возрастную категорию для игры. Описание содержимого располагается на обратной стороне коробки, обычно в нижнем левом или правом углу и перечисляет некоторые элементы содержания игры.

В нашем датафрейме выделены следующие категории с частотой по убыванию:

- «E» («Everyone») — «Для всех»: Содержание вполне подходит для всех возрастов. Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера. Первой игрой, которая получила данный рейтинг, стала The Simpsons Cartoon Studio, выпущенная в 1996 году. Первоначально "K-A" ("Kids to Adults").

- «T» («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, в меру откровенное сексуальное содержимое, кровь или нечастое использование ненормативной лексики.

- «M» («Mature») — «Для взрослых»: Материалы игры не подходят для лиц младше 17 лет. Проекты с данным рейтингом могут содержать достаточно жестокое насилие, большое количество крови с расчленением, непристойные сексуальные сцены или грубую ненормативную лексику, нежелательную для младшей аудитории.

## Этап 5. Проверка гипотез

Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;
Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.
Задайте самостоятельно пороговое значение alpha.
Поясните:
 - Как вы сформулировали нулевую и альтернативную гипотезы;
 - Какой критерий применили для проверки гипотез и почему.

```python
# выделим пользовательские рейтинги по платформе XOne  и рассчитаем среднее значение

xone = data.loc[(data['platform'] == 'XOne') & (data['user_score'] > 0), 'user_score']
xone.mean()
```

```python
# выделим пользовательские рейтинги по платформе PC  и рассчитаем среднее значение

pc = data.loc[(data['platform'] == 'PC') & (data['user_score'] > 0), 'user_score']
pc.mean()
```

Как видно из расчетов, средние пользовательские рейтинги по платформам XboxOne и PC разные. Проверим нашу гипотезу специальным статистическим методом **ttest_ind** для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам.

Для проверки гипотезы "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые" сформулируем нулевую и альтернативную гипотезы:

H_0: Средние пользовательские оценки (user_score) платформы "XOne" = Средние пользовательские оценки (user_score) платформы "PC"

H_a: Средние пользовательские оценки (user_score) платформы "XOne" ≠ Средние пользовательские оценки (user_score) платформы "PC"

Установим критический уровень статистической значимости alpha = 0.05.

```python
# проверим методом ttest_ind нашу нулевую гипотезу

results = st.ttest_ind(
    xone, 
    pc, equal_var = False)

alpha = 0.05 

# выведем значение p-value на экран 
print(results.pvalue) 
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

p-value - это вероятность получить наблюдаемый или еще более далекий от предполагаемого результат при условии, что нулевая гипотеза верна. Поскольку полученное значение p-value значительно меньше заданного уровня статистической значимости, то мы отвергаем нашу нулевую гипотезу о том, что средние пользовательские оценки платформы 'XOne' и 'PC' одинаковые и принимаем верной альтернативную гипотезу о том, что средние пользовательские оценки платформы 'XOne' и 'PC' разные.

```python
# выделим пользовательские рейтинги жанра игр Action и рассчитаем среднее значение

action = data.loc[(data['genre'] == 'Action') & (data['user_score'] > 0), 'user_score']
action.mean()
```

```python
# выделим пользовательские рейтинги жанра игр Sports и рассчитаем среднее значение

sports = data.loc[(data['genre'] == 'Sports') & (data['user_score'] > 0), 'user_score']
sports.mean()
```

Как видно из расчетов, средние пользовательские рейтинги игровых жанров Action и Sports разные. Проверим нашу гипотезу специальным статистическим методом **ttest_ind** для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам.

Для проверки гипотезы "Средние пользовательские рейтинги игровых жанров Action и Sports разные" сформулируем нулевую и альтернативную гипотезы:

H_0: Средние пользовательские оценки (user_score) жанра "Action" = Средние пользовательские оценки (user_score) жанра "Sports"

H_a: Средние пользовательские оценки (user_score) жанра "Action" ≠ Средние пользовательские оценки (user_score) жанра "Sports"

Установим критический уровень статистической значимости alpha = 0.05.

```python
# проверим методом ttest_ind нашу нулевую гипотезу

results = st.ttest_ind(
    action, 
    sports, equal_var = False)

alpha = 0.05 

# выведем значение p-value на экран 
print(results.pvalue) 
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Поскольку полученное значение p-value получилось больше заданного уровня статистической значимости, то мы не отвергаем нашу нулевую гипотезу о том, что средние пользовательские оценки игровых жанров Action и Sports одинаковые.
