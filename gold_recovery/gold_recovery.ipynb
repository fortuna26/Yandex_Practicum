### Этап 1.2 Проверка расчета эффективности обогащения

Проверим, что эффективность обогащения рассчитана правильно. Вычислим её на **обучающей выборке** для признака **rougher.output.recovery.** Найдем **MAE** между получившимися расчётами и значением признака. Сделаем вывод.

Эффективность обогащения рассчитывается по формуле:



где:
 * C — доля золота в концентрате после флотации/очистки;
 * F — доля золота в сырье/концентрате до флотации/очистки;
 * T — доля золота в отвальных хвостах после флотации/очистки.

```python
# обозначим переменные для расчета recovery

C = gold_rec_train['rougher.output.concentrate_au']
F = gold_rec_train['rougher.input.feed_au']
T = gold_rec_train['rougher.output.tail_au']
```

```python
# посчитаем recovery

gold_rec_train['recovery'] = ((C * (F - T)) / (F * (C - T))) * 100
gold_rec_train['recovery'].mean()
```

```python
# найдем среднее значение rougher.output.recovery в обучающей выборке

gold_rec_train['rougher.output.recovery'].mean()
```

Эффективность обогащения рассчитана правильно.

```python
# рассчитаем среднюю абсолютную ошибку между нашими расчетами и значениями признака

mae = mean_absolute_error(gold_rec_train['recovery'], gold_rec_train['rougher.output.recovery'])
mae
```

Получилось супермаленькое значение **mae** между нашими расчетами и значениями признака, и значит эффективность обогащения рассчитана верно.

### Этап 2.1 Изучение концентрации металлов (Au, Ag, Pb) на различных этапах очистки

```python
# для удобства отбора этапов технологического процесса и концентрации металлов выведем на экран все столбцы обучающей выборки

gold_rec_train.columns
```

```python
# создадим табличку из данных по концентрации золота Au по этапам очистки

content_au = gold_rec_train[['rougher.input.feed_au',
                             'rougher.output.concentrate_au',
                             'primary_cleaner.output.concentrate_au',
                             'final.output.concentrate_au']].rename(columns={
    'rougher.input.feed_au': 'исходное сырье Au',
    'rougher.output.concentrate_au': 'флотация Au',
    'primary_cleaner.output.concentrate_au':'первичная очистка Au',
    'final.output.concentrate_au': 'финальная очистка Au'})

content_au.head()
```

Вполне логично, что концентрация золота к финальной очистке стремительно растет.

```python
# создадим табличку из данных по концентрации серебра Ag по этапам очистки

content_ag = gold_rec_train[['rougher.input.feed_ag',
                             'rougher.output.concentrate_ag',
                             'primary_cleaner.output.concentrate_ag',
                             'final.output.concentrate_ag']].rename(columns={
    'rougher.input.feed_ag': 'исходное сырье Ag',
    'rougher.output.concentrate_ag': 'флотация Ag',
    'primary_cleaner.output.concentrate_ag':'первичная очистка Ag',
    'final.output.concentrate_ag': 'финальная очистка Ag'})

content_ag.head()
```

Концентрация серебра во время флотации растёт, затем к финальной очистке падает.

```python
# создадим табличку из данных по концентрации свинца Pb по этапам очистки

content_pb = gold_rec_train[['rougher.input.feed_pb',
                             'rougher.output.concentrate_pb',
                             'primary_cleaner.output.concentrate_pb',
                             'final.output.concentrate_pb']].rename(columns={
    'rougher.input.feed_pb': 'исходное сырье Pb',
    'rougher.output.concentrate_pb': 'флотация Pb',
    'primary_cleaner.output.concentrate_pb':'первичная очистка Pb',
    'final.output.concentrate_pb': 'финальная очистка Pb'})

content_pb.head()
```

Концентрация свинца ведет себя нестабильно от стадии к стадии: исходное сырье свинца во время флотации и на первичной очистке растет, но к финальной очистке концентрация может как незначительно упасть, так и незначительно вырасти.

Посмотрим на графике распределение средних значений коцентраций металлов по стадиям

```python
# построим графики средней концентрации веществ на разных стадиях очистки

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, linewidth=3, figsize=(25, 7))

content_au.mean().plot(kind='bar', ax=ax1, fontsize=12, grid=True, color='yellow')
ax1.set_title('Золото (Au)', fontsize=16)
ax1.set_ylabel('Концентрация золота Au', fontsize=14)
ax1.set_xlabel('Стадии очистки', fontsize=14)
ax1.set_xticklabels(ax1.get_xticklabels(), rotation = 30)

content_ag.mean().plot(kind='bar', ax=ax2, fontsize=12, grid=True, color='grey')
ax2.set_title('Серебро (Ag)', fontsize=16)
ax2.set_ylabel('Концентрация серебра Ag', fontsize=14)
ax2.set_xlabel('Стадии очистки', fontsize=14)
ax2.set_xticklabels(ax2.get_xticklabels(), rotation = 30)

content_pb.mean().plot(kind='bar', ax=ax3, fontsize=12, grid=True, color='black')
ax3.set_title('Свинец (Pb)', fontsize=16)
ax3.set_ylabel('Концентрация свинца Pb', fontsize=14)
ax3.set_xlabel('Стадии очистки', fontsize=14)
ax3.set_xticklabels(ax3.get_xticklabels(), rotation = 30)

plt.suptitle('Изменение концентрации металлов (Au, Ag, Pb) на различных этапах очистки', fontsize=20);
```

Как и было отмечено ранее, к финальной очистке концентрация золота - растёт, серебра - падает. Концентрация свинца от исходного сырья к флотации растет стремительно, затем, к финальной очистке - незначительно. 

Во время флотации концентрации всех металлов растут.

## Этап 3. Построение модели

### Этап 3.1 Метрики оценки качества

Вместо МАЕ введём новую метрику качества — sMAPE (англ. Symmetric Mean Absolute Percentage Error, «симметричное среднее абсолютное процентное отклонение»). Она похожа на MAE, но выражается не в абсолютных величинах, а в относительных, и одинаково учитывает масштаб и целевого признака, и предсказания.
Метрика sMAPE вычисляется так:




где:

 * $y_i$ - значение целевого признака для объекта с порядковым номером i в выборке, на которой измеряется качество
 * $\hat{y}_i$ - значение предсказания для объекта с порядковым номером i, например, в тестовой выборке
 * N - количество объектов в выборке
 * $\sum_{i=1}^{N}$ - суммирование по всем объектам выборки (i меняется от 1 до N)

Нужно спрогнозировать сразу две величины:

 * эффективность обогащения чернового концентрата rougher.output.recovery;
 * эффективность обогащения финального концентрата final.output.recovery.

Итоговая метрика складывается из двух величин:



### Этап 3.2 Функция для вычисления итоговой sMAPE

```python
# напишем функции для вычисления метрик sMAPE

def smape(target, predict):
    return (((abs(target - predict)) / ((abs(target) + abs(predict)) / 2)).sum() / len(target)) * 100

def total_smape(smape_rougher, smape_final):
    return 0.25 * smape_rougher.item() + 0.75 * smape_final.item()
```

### Этап 3.4 Обучение моделей и оценка их качества кросс-валидацией

```python
# укажем, что метрику нужно минимизировать

smape_score = make_scorer(smape, greater_is_better=False)
```

```python
# обучим модель на алгоритме линейной регрессии

model_lr = LinearRegression()
smape_rougher = np.mean(cross_val_score(model_lr, features_train_rougher, target_train_rougher, 
                                        scoring=smape_score, cv=5, n_jobs=2))
smape_final = np.mean(cross_val_score(model_lr, features_train_final, target_train_final, 
                                      scoring=smape_score, cv=5, n_jobs=2))

print(smape_rougher, smape_final)
print('Итоговое sMAPE модели линейной регрессии:', total_smape(smape_rougher, smape_final))
```

```python
# обучим модель на алгоритме решающего дерева

for depth in range(1, 11):
    model_dtr = DecisionTreeRegressor(random_state=random_state, max_depth = depth)
    smape_rougher = np.mean(cross_val_score(model_dtr, features_train_rougher, target_train_rougher, 
                                            scoring=smape_score, cv=5, n_jobs=2))
    smape_final = np.mean(cross_val_score(model_dtr, features_train_final, target_train_final, 
                                          scoring=smape_score, cv=5, n_jobs=2))

    print('Глубина:', depth)
    print(smape_rougher, smape_final)
    print('Итоговое sMAPE модели решающего дерева:', total_smape(smape_rougher, smape_final))
    print('-'*25)
```

У модели решающего дерева при глубине depth= 3 получилось наименьшее итоговое sMAPE = -8.4622787502315

```python
# обучим модель на алгоритме случайного леса

best_model_rfr = None
best_smape = 0

for est in range(10, 51, 10):
    model_rfr = RandomForestRegressor(random_state=random_state, max_depth = 3, n_estimators = est)
    smape_rougher = np.mean(cross_val_score(model_rfr, features_train_rougher, target_train_rougher, 
                                                scoring=smape_score, cv=5, n_jobs=2))
    smape_final = np.mean(cross_val_score(model_rfr, features_train_final, target_train_final, 
                                              scoring=smape_score, cv=5, n_jobs=2))
       
    total_smape = 0.25 * smape_rougher.item() + 0.75 * smape_final.item()
        
    if total_smape < best_smape:
        best_model_rfr = model_rfr 
        best_smape = total_smape 
        
print('Итоговое sMAPE модели случайного леса:', best_smape)
print('Лучшая модель:', best_model_rfr)
```

Поскольку лучшее sMAPE показала модель случайного леса = -8.279356109669802 при количестве деревьев = 20, то именно её мы и будем тестировать на тестовом датафрейме.
