_**Цель проекта: разработать модель для прогнозирования температуры стали. Добиться значения метрики MAE <= 6.8.**_

_**Описание проекта:** Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали._

_Обработка стали происходит следующим образом: расплавленную сталь заливают в ковш, облицованный огнеупорным кирпичом, и подогревают до нужной температуры графитовыми электродами, которые установлены в крышке ковша. Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Затем сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»)._

_Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки._

_Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. *slab*, «плита»)._

_Построенная модель предскажет конечную температуру стали. При управлении нагревом можно будет подставлять разные параметры и корректировать параметры так, чтобы избегать лишнего нагрева. Можно сказать, что с помощью модели будет происходить имитация технологического процесса. Т.е. мы сможем воспроизводить процесс(эксперимент) с помощью математической модели, и тестировать только лучшие результаты, также можем выделить наиболее важные признаки влияющие на нагрев и конечную температуру._

Добавим признаков в этой таблице из имеющихся. Тем более так как соединять данные надо будет по полю key, то надо сделать номер партии уникальным, сгруппировав все данные по номеру партии.

Посчитаем полную мощность из активной и реактивной; посчитаем длительность каждого нагрева и заодно избавимся от временных меток; затем сгруппируем по номеру партии и выведем суммарную информацию по затраченной мощности, длительности нагрева и количеству итераций нагрева.

```python
# добавим столбец с полной мощностью с помощью функции

def full_power(row):
    full_power = m.sqrt(row['Активная мощность']**2 + row['Реактивная мощность']**2)
    return full_power

data_arc['full_power'] = data_arc.apply(full_power, axis=1)
```

```python
# проверим, что получилось

data_arc.head()
```

```python
# переведем столбцы с датой и временем в тип datetime, чтобы посчитать временную разницу

data_arc['Начало нагрева дугой'] = data_arc['Начало нагрева дугой'].astype('datetime64[ns]')
data_arc['Конец нагрева дугой'] = data_arc['Конец нагрева дугой'].astype('datetime64[ns]')
```

```python
# добавим новый столбец с длительностью нагрева на каждой итерации

data_arc['duration'] = (data_arc['Конец нагрева дугой'] - data_arc['Начало нагрева дугой'])
```

```python
# переведем новый столбец с получившимся типом данных timedelta в минуты (числовой тип)

data_arc['duration'] = data_arc['duration'] / pd.Timedelta(minutes=1)
```

```python
# проверим, что получилось

data_arc.head()
```

```python
# сгруппируем данные по уникальным партиям, заодно сразу переименуем столбцы с рус.яз.на англ.яз. в змеином стиле


data_arc_features = data_arc.groupby('key', as_index=False).agg({'Активная мощность': 'sum',
                                                                 'Реактивная мощность': 'sum',
                                                                 'full_power': 'sum', 
                                                                 'duration': 'sum',
                                                                 'Начало нагрева дугой': 'count'}).\
                                             rename(columns={'Активная мощность': 'sum_activ_power',
                                                             'Реактивная мощность': 'sum_reactiv_power',
                                                             'full_power': 'sum_full_power',
                                                            'duration': 'sum_duration',
                                                            'Начало нагрева дугой': 'iterations'})
```

Завод по производству стали ООО "Так закаляем сталь" (Заказчик) обратился для разработки некоего алгоритма, который помог бы оптимизировать производство, и в первую очередь экономить электроэнергию.

Завод использует современную технологию производства стали с помощью дуговой электропечи. Дуговая электропечь по принципу действия является мартеновской, но нагревание расплава в ней происходит за счёт тепла электрической дуги на графитовых электродах.





Сталь обрабатывают в металлическом ковше, облицованным огнеупорным кирпичом, вместимостью около 100 тонн. Расплавленную сталь заливают в ковш и подогревают до нужной температуры, из сплава выводится сера, добавляются примеси, корректирующие химический состав и отбираются пробы. Затем сталь легируют - добавляют различные сыпучие и проволочные материалы, всё это постоянно перемешивают продуванием инертным газом.

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. После легирования снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Необходимо, чтобы завод с помощью математической модели помимо предсказания конечной температуры стали мог бы воспроизводить математический эксперимент без самого физического процесса, научиться управлять процессом производства стали, сократить количество итераций нагрева (тем самым электроэнергию) и длительность цикла, быстрее достигая целевого химического состава и нужную температуру плавки.
